```python
# Importing necessary libraries
import numpy as np  # For numerical operations
import pandas as pd  # For data manipulation and processing
import matplotlib.pyplot as plt  # For plotting graphs
import matplotlib.dates as mdates  # For formatting dates in plots
import seaborn as sns  # For advanced visualizations
from sklearn.preprocessing import MinMaxScaler  # For scaling data
from tensorflow.keras.models import Sequential  # For building neural network models
from tensorflow.keras.layers import LSTM, Dense, Dropout  # LSTM and other layers for neural network

# Configuring warnings to ignore them (to keep the notebook clean)
import warnings
warnings.filterwarnings('ignore')

# Setting seaborn for better visualization aesthetics
colors = ['#FF9900', '#000000']
sns.set(palette=colors, font='Serif', style='white', rc={'axes.facecolor': 'whitesmoke', 'figure.facecolor': 'whitesmoke'})
```

```python
# Loading the dataset
df1 = pd.read_csv('TSLA.csv')

# Displaying the first few rows of the dataset for an overview
print(df1.head())

# Displaying summary statistics of the dataset
print(df1.describe())

# Calculating and displaying the median for numeric columns, rounded to two decimal places
medians = df1.median(numeric_only=True).round(2)
print("Medians:\n", medians)

# Converting 'Date' column to datetime and sorting the DataFrame by date
df1['Date'] = pd.to_datetime(df1['Date'])
df1.sort_values('Date', inplace=True)

# Displaying the minimum and maximum dates to understand the time range of the dataset
print("Minimum date value:", df1['Date'].min())
print("Maximum date value:", df1['Date'].max())
```

```python
# Visualizing the Open and Close Prices of Tesla Stock

fig = plt.figure(figsize=(20, 8))
plt.title("TESLA Stock Value Changes Over Time", size=20, weight='bold')

# Plotting Open Prices
ax1 = sns.lineplot(data=df1, x='Date', y='Open', label='Open Price', color=colors[0])
# Plotting Close Prices
ax2 = sns.lineplot(data=df1, x='Date', y='Close', label='Close Price', color=colors[1])

# Customizing the plot aesthetics
for spine in ['left', 'right', 'top', 'bottom']:
    ax1.spines[spine].set_visible(False)

plt.legend()
plt.show()

# Visualizing the Trading Volume of Tesla Stock

fig = plt.figure(figsize=(20, 8))
plt.title("TESLA Stock Trading Volume Over Time", size=20, weight='bold')

# Plotting Volume
ax3 = sns.lineplot(data=df1, x='Date', y='Volume', color=colors[0])

# Customizing the plot aesthetics
for spine in ['left', 'right', 'top', 'bottom']:
    ax3.spines[spine].set_visible(False)

plt.show()
```

```python
# Setting up the figure for distribution plots
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
fig.suptitle("Analyzing the Numeric Variables - Distribution and Box Plots", size=20, weight='bold')

# Creating distribution plots for 'High', 'Low', 'Open', 'Close', and 'Volume'
sns.kdeplot(df1['High'], ax=axes[0, 0], color=colors[0], fill=True, linewidth=2)
sns.kdeplot(df1['Low'], ax=axes[0, 1], color=colors[1], fill=True, linewidth=2)
sns.kdeplot(df1['Open'], ax=axes[0, 2], color=colors[0], fill=True, linewidth=2)
sns.kdeplot(df1['Close'], ax=axes[1, 0], color=colors[1], fill=True, linewidth=2)
sns.kdeplot(df1['Volume'], ax=axes[1, 1], color=colors[0], fill=True, linewidth=2)

# Customizing the plot aesthetics
for ax in axes.flat:
    ax.label_outer()
    for spine in ['left', 'right', 'top', 'bottom']:
        ax.spines[spine].set_visible(False)

# Setting up the figure for box plots
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
fig.suptitle("Boxplot of Numeric Variables", size=20, weight='bold')

# Creating box plots for 'High', 'Low', 'Open', 'Close', and 'Volume'
sns.boxplot(df1['High'], ax=axes[0, 0], color=colors[0])
sns.boxplot(df1['Low'], ax=axes[0, 1], color=colors[1])
sns.boxplot(df1['Open'], ax=axes[0, 2], color=colors[0])
sns.boxplot(df1['Close'], ax=axes[1, 0], color=colors[1])
sns.boxplot(df1['Volume'], ax=axes[1, 1], color=colors[0])

# Customizing the plot aesthetics
for ax in axes.flat:
    ax.label_outer()
    for spine in ['left', 'right', 'top', 'bottom']:
        ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# Creating a pair plot to visualize pairwise relationships in the dataset
sns.pairplot(df1, corner=True)
plt.suptitle("Pairwise Relationships Among Features", size=20, weight='bold')
plt.show()

# Analyzing the correlation of different features with the 'Close' price
correlation_with_close = df1.corr()['Close'].sort_values(ascending=False)
print("Correlation with 'Close' Price:\n", correlation_with_close)
```

```python
from scipy.stats import shapiro

# Selecting numeric columns for normality testing
int_cols = df1.select_dtypes(include=['int', 'float']).columns.to_list()

# Performing Shapiro-Wilk normality test on each numeric feature
for feature in int_cols:
    _, p_value = shapiro(df1[feature])
    if p_value < 0.05:
        print(f"Feature {feature} is not normally distributed (p-value: {p_value:.3f})")
    else:
        print(f"Feature {feature} is normally distributed (p-value: {p_value:.3f})")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Selecting features and target variable for the Linear Regression model
X = df1[['Volume', 'Open']]  # Features
y = df1['Close']  # Target variable

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False, random_state=42)

# Scaling the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initializing and training the Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train_scaled, y_train)

# Making predictions
predictions = linear_model.predict(X_test_scaled)

# Evaluating the model
r2 = r2_score(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)

# Displaying the evaluation metrics
print("Linear Regression Model Evaluation:")
print(f"R2 Score: {r2:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
```

```python
# Plotting for Linear Regression Model
plt.figure(figsize=(15, 6))
plt.plot(y_test.index, y_test, label='Actual Values', color='blue')
plt.plot(y_test.index, predictions, label='Predicted Values', color='red')
plt.title('Linear Regression: Actual vs Predicted')
plt.xlabel('Index')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
# Preprocessing data for LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_close = scaler.fit_transform(df1['Close'].values.reshape(-1, 1))

def create_dataset(data, time_step=1):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):
        a = data[i:(i + time_step), 0]
        X.append(a)
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

time_step = 10
X, y = create_dataset(scaled_close, time_step)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Splitting data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Building the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compiling the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Training the model
model.fit(X_train, y_train, batch_size=1, epochs=10)

# Making predictions
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Inverse transforming predictions
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)

# Calculating RMSE performance metrics
train_rmse = np.sqrt(np.mean((train_predict - y_train.reshape(-1, 1)) ** 2))
test_rmse = np.sqrt(np.mean((test_predict - y_test.reshape(-1, 1)) ** 2))

# Displaying the performance metrics
print("LSTM Model Performance:")
print(f"Train RMSE: {train_rmse:.2f}")
print(f"Test RMSE: {test_rmse:.2f}")
```

```python
# Inverse transform the scaled 'Close' prices to get the actual prices
actual_prices = scaler.inverse_transform(scaled_close).reshape(-1)

# Plotting for LSTM Model
plt.figure(figsize=(15, 6))

# Plotting the actual prices
plt.plot(df1['Date'], actual_prices, label='Actual Prices', color='blue')

# Plotting the train predictions
plt.plot(df1['Date'][:len(train_predict)], train_predict, label='Train Predictions', color='green')

# Plotting the test predictions
test_dates = df1['Date'][len(train_predict):len(train_predict) + len(test_predict)]
plt.plot(test_dates, test_predict, label='Test Predictions', color='red')

plt.title('LSTM: Actual and Predicted Prices')
plt.xlabel('Date')
plt.ylabel('Prices')
plt.legend()
plt.show()
```

```python
from prophet import Prophet
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Preparing the dataset for Prophet
df_p = df1[['Date', 'Close']]
df_p.columns = ['ds', 'y']

# Splitting the dataset into training and testing sets
split_percentage = 0.73
split_index = int(len(df_p) * split_percentage)
train = df_p[:split_index]
test = df_p[split_index:]

# Creating and fitting the Prophet model
prophet_model = Prophet()
prophet_model.fit(train)

# Creating a future DataFrame for the entire period
future = prophet_model.make_future_dataframe(periods=len(test))

# Making predictions
forecast = prophet_model.predict(future)

# Merging the forecast with the original dataset
merged_data = df_p.merge(forecast[['ds', 'yhat']], on='ds', how='outer')

# Handling NaN values in the 'yhat' column (predictions)
# If there are any NaN values, you might consider forward filling (ffill)
merged_data['yhat'].fillna(method='ffill', inplace=True)

# If the original 'y' (Close prices) has NaNs, handle them as well
merged_data['y'].fillna(method='ffill', inplace=True)

# Splitting merged data into train and test for evaluation
train_data = merged_data[:split_index]
test_data = merged_data[split_index:]

# Evaluating the model on the test set
r2 = r2_score(test_data['y'], test_data['yhat'])
mse = mean_squared_error(test_data['y'], test_data['yhat'])
mae = mean_absolute_error(test_data['y'], test_data['yhat'])

# Displaying the evaluation metrics
print("Prophet Model Evaluation:")
print(f"R2 Score: {r2:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")

```

```python
# Plotting the Prophet model's predictions against the actual values
plt.figure(figsize=(15, 8))
plt.plot(merged_data['ds'], merged_data['y'], label='Actual Values', color='blue')
plt.plot(merged_data['ds'], merged_data['yhat'], label='Predicted Values', color='red')

# Optionally, if 'yhat_lower' and 'yhat_upper' are available in the forecast,
# you can include a fill to show the uncertainty interval
# plt.fill_between(merged_data['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='pink', alpha=0.5)

plt.title('Prophet Model: Actual vs Predicted')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
